```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# تنظیمات اولیه
tspan = np.array([0, 100])
x0 = np.array([0, 20, 20, 20, 20, 0, 0, 0, 0, 0])

# تابع مدل قطار برای معادلات دیفرانسیل
def train_model(x, t):
    # پارامترها
    F = 1000  # مقدار فرضی برای پارامتر F

    # وضعیت‌های مدل
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    x5 = x[4]
    v1 = x[5]
    v2 = x[6]
    v3 = x[7]
    v4 = x[8]

    # معادلات دیفرانسیل
    dxdt = np.zeros_like(x)
    dxdt[0] = x2  # dx1/dt = x2
    dxdt[1] = x3  # dx2/dt = x3
    dxdt[2] = x4  # dx3/dt = x4
    dxdt[3] = x5  # dx4/dt = x5
    dxdt[4] = F   # dx5/dt = F
    dxdt[5] = 0   # dv1/dt = 0
    dxdt[6] = 0   # dv2/dt = 0
    dxdt[7] = 0   # dv3/dt = 0
    dxdt[8] = 0   # dv4/dt = 0

    return dxdt

# حل معادلات دیفرانسیل با استفاده از odeint
t = np.linspace(tspan[0], tspan[1], 1000)  # تعداد بیشتری از نقاط برای دقت بالاتر
x = odeint(train_model, x0, t)

# رسم نمودارها
plt.figure(figsize=(10, 8))

plt.subplot(211)
plt.plot(t, x[:, 0], 'k')
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('Locomotive Position (m)')
plt.legend(['x_1'])

plt.subplot(212)
plt.plot(t, x[:, 1], 'k', label='x_2')
plt.plot(t, x[:, 4], 'k-.', label='x_5')
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('Wagons Distance (m)')
plt.legend()

plt.tight_layout()
plt.show()
```
