```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def inverted_pendulum_k2(t, x):
    # Parameters (adjust these according to your system)
    m = 0.5  # Mass of the cart
    M = 0.2  # Mass of the pendulum
    L = 0.3  # Length of the pendulum
    g = 9.81  # Acceleration due to gravity
    k2 = 0.5  # Some parameter for the dynamics

    # State variables
    dxdt = np.zeros_like(x)
    dxdt[0] = x[1]
    dxdt[1] = (1/(m + M - m*np.cos(x[2])**2)) * (k2*x[3] - m*L*x[3]**2*np.sin(x[2])*np.cos(x[2]) - (m + M)*g*np.sin(x[2]))
    dxdt[2] = x[3]
    dxdt[3] = (1/(L*(m + M - m*np.cos(x[2])**2))) * ((m*L*x[3]**2*np.sin(x[2])*np.cos(x[2]) - (m + M)*g*np.sin(x[2]))*np.cos(x[2]) - k2*x[1])

    return dxdt

# Initial conditions
x0 = np.array([0, 0, 0.6, 0])  # [x (m), dx/dt (m/s), theta (rad), dtheta/dt (rad/s)]

# Time span
tspan = (0, 3)

# Solve the ODE
sol = solve_ivp(inverted_pendulum_k2, tspan, x0, method='RK45', max_step=1e-2)

# Extract results
t = sol.t
x_x = sol.y[0]
x_theta = sol.y[2]

# Plotting
plt.figure()
plt.plot(t, x_x, 'k', label='x (m)', linewidth=2)
plt.plot(t, x_theta, '-.k', label='theta (rad)', linewidth=2)
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('State Variables')
plt.legend()
plt.title('Simulation of Inverted Pendulum System')
plt.show()
```
