```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# تابع برای مدل مخزن
def tank_model(t, x):
    A = 1.0
    C = 2.0
    F_in = 0.0  # بدون ورودی اختلال
    u = 0.1  # باز شدن ثابت برای دریچه

    dxdt = 1/A * (F_in - C * u * np.sqrt(x))
    return dxdt

# پارامترهای شبیه‌سازی
tspan = (0, 10)
y0 = [1.0]

# حل معادلات دیفرانسیل
sol = solve_ivp(tank_model, tspan, y0, max_step=1e-2)

# رسم نتایج شبیه‌سازی
plt.plot(sol.t, sol.y[0], 'k', label='Tank Level (l)')
plt.grid()
plt.xlabel('Time (sec)')
plt.ylabel('Tank Level')
plt.legend()
plt.gca().lines[0].set_linewidth(2)
plt.show()
```
